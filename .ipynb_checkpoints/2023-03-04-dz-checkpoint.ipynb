### Календарь

```python
def is_leap(year) -> bool:
    return year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)

def get_month_days(month : int, leap : bool) -> int:
    if month in (4, 6, 11):
        return 30
    elif (month == 2) and (leap == 1):
        return 29
    elif (month == 2) and (leap == 0):
        return 28
    else:
        return 31

    
#Воспользуемся алгоритмом Rata Die
def get_first_weekday(month : int, year : int) -> int:  #результат 1 .. 6
    D = 6 #01.01.2000 - суббота
    weekday = (D + get_num_of_days_since_20000101(month, year)) % 7
    
    return weekday - 1

def get_num_of_days_since_20000101(month : int, year : int) -> int:
    total_days = 0
    for y in range(2000, year):
        if is_leap(y):
            total_days += 366
        else:
            total_days += 365
    for m in range(1, month):
        total_days += get_month_days(m, is_leap(year))
        
    return total_days

def format_workday(day = None, sep = ' ') -> str:
    day = day or ''
    s = f" {day:<2}" + sep 
    return s

def format_holiday(day = None, sep = ' ') -> str:
    blue = "34m"
    day = day or ''
    s = f" \033[{blue}{day}\033[00m" +  sep
    return s


def is_work_day(i):
    return i % 7 <= 5

def print_days(month_days, first_weekday, sep = ' '):
    w_days = ['Mon', 'Tue', 'Wed','Thu', 'Fri', 'Sat', 'Sun']
    s = sep.join(w_days) + '\n'

    for i in range(first_weekday + month_days):
        if i < first_weekday:
            day = None
        else:
            day = i - first_weekday + 1
            
        if is_work_day(i):
            s += format_workday(day, sep)
        else:
            s += format_holiday(day, sep)
        if i % 7 == 6:
            s += '\n' 
    return s

def calendar (month, year):
    month_days = get_month_days(month, is_leap(year))
    first_weekday = get_first_weekday(month, year)
    return print_days(month_days, first_weekday, sep = '  ')
```

```python
print(calendar (2, 2023))
```

```python
[print(i) for i in zip(calendar (1, 2023), calendar (2, 2023))]
```

```python
list( zip(calendar (1, 2023), calendar (2, 2023)))
```

### Разбить строку по разделителю

```python
s = 'Hdfs  Tutorial,,'
```

```python
def split_string(s, sep = ' ,;.'):
    i = 0
    words = []
    word = ''
    n = len(s) - 1  

#     if s[n] in sep:   #если последний знак - разделитель, сдвигаемся на один индекс
#         n = len(s) - 2

    while i <= n:
        if s[i] not in sep:
            word += s[i]
            i += 1
        else:
            if word:
                words.append(word)
            word = ''
            i += 1
    if word:
        words.append(word)
    
    return words
```
